### NORMALISATION

```python
train_dataNormalized=data_train2.copy()
test_dataNormalized=data_test2.copy()
dataset = pd.concat([data_train2, data_test2], ignore_index=True)
def normalize(df, cols):
    result = df.copy() # do not touch the original df
    for feature_name in cols:
        if feature_name!='xAttack':
            max_value = dataset[feature_name].max()
            min_value = dataset[feature_name].min()
            if max_value > min_value:
                result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

#Normalizing training set
train_dataNormalized = normalize(train_dataNormalized,data_train2.columns[:28])
test_dataNormalized = normalize(test_dataNormalized,data_test2.columns[:28])
```

### Feature Gaussian Mixture Probability Model 

```python
from sklearn.mixture import GaussianMixture
from scipy.stats import norm
from scipy import stats
from scipy.io import loadmat
```

```python
y_binary_train = y_train.apply(lambda x: 0 if x == 0 else 1)
y_binary_test = y_test.apply(lambda x: 0 if x == 0 else 1)
```

```python
filtered_data = d_norm[y_binary_train == 0]
n_components = 1  # You can set the number of components as needed
gmm = GaussianMixture(n_components=n_components, random_state=42)  # Using a seed for reproducibility
gmm.fit(filtered_data)

# Get the means and covariances of the Gaussian components
means = gmm.means_
covariances = gmm.covariances_

print(covariances[0][0][1])
print(covariances[0][1][0])
```

```python
def feature_probs(df, df_test, y):
    result = df.copy()
    result_test = df_test.copy()
    normal_trafic = result[y == 0]
    # print(normal_trafic.shape)
    n_components = 1  # You can set the number of components as needed
    gmm = GaussianMixture(n_components=n_components, random_state=42)  # Using a seed for reproducibility
    gmm.fit(normal_trafic)

    # Get the means and covariances of the Gaussian components
    means = gmm.means_[0]
    covariances = gmm.covariances_
    
    variances = []
    for i in range(0,27):
        variances.append(covariances[0][i][i])

    std_dev = np.sqrt(variances)  # Calculating standard deviation from variances

    for idx, feature in enumerate(result.columns):
        mean = means[idx]
        std = std_dev[idx]

        if isinstance(mean, pd.Series):
            mean = mean.values  # Extracting numpy array if mean is a pandas Series

        if isinstance(std, pd.Series):
            std = std.values  # Extracting numpy array if std_dev is a pandas Series

        if np.isscalar(mean) and np.isscalar(std):
            if std > 0:  # Ensuring standard deviation is non-zero
                mask = result[feature] > mean
                result.loc[mask, feature] = 2 * (1 - norm.cdf(result.loc[mask, feature], loc=mean, scale=std))
                result.loc[~mask, feature] = 2 * norm.cdf(result.loc[~mask, feature], loc=mean, scale=std)
                
                mask_test = result_test[feature] > mean
                result_test.loc[mask_test, feature] = 2 * (1 - norm.cdf(result_test.loc[mask_test, feature], loc=mean, scale=std))
                result_test.loc[~mask_test, feature] = 2 * norm.cdf(result_test.loc[~mask_test, feature], loc=mean, scale=std)
            else:
                print(f"Skipping {feature} due to standard deviation of zero.")

    return result, result_test
```

```python
d_raw_probs, d_raw_probs_test = feature_probs(d_raw, d_raw_test, y_binary_train)
d_raw_pca_probs, d_raw_pca_probs_test = feature_probs(d_raw_pca, d_raw_pca_test, y_binary_train)
d_norm_probs, d_norm_probs_test = feature_probs(d_norm, d_norm_test, y_binary_train)
d_norm_pca_probs, d_norm_pca_probs_test = feature_probs(d_norm_pca, d_norm_pca_test, y_binary_train)
```

## Voting algorithm:

```python
def classify_traffic(vector, threshold, consensus):
    positive_votes = 0
    
    # Parcourir chaque caractéristique dans le vecteur de trafic
    for feature_prob in vector:
        # Vérifier si la probabilité de la caractéristique est supérieure au seuil individuel
        if feature_prob < threshold:
            positive_votes += 1
            
    
    # Classer le vecteur de trafic en fonction du nombre de votes positifs
    if positive_votes >= consensus:
        return 1  # Considérer comme anormal
    else:
        return 0     # Considérer comme normal
```

## K-Means Anomoly Detection:

```python
from cuml.cluster import KMeans
```

```python
def detect_anomalies_with_kmeans(data, y, test, anomaly_threshold=1):
    # Fit K-means model on normal data
    kmeans = KMeans(n_clusters=2)  # Assuming normal data forms 2 clusters
    normal_data =  data[y == 0]
    kmeans.fit(normal_data)

    # Calculate distances to centroids
    distances = kmeans.transform(test)
    centroid_distances = np.min(distances, axis=1)
    # print(centroid_distances)

    # Define anomalies based on distance threshold from centroids
    anomalies = centroid_distances > anomaly_threshold  # Adjust threshold as needed

    # Convert boolean anomalies to 0s and 1s
    anomaly_labels = anomalies.astype(int)

    return anomaly_labels

def plot_f1_scores(data, y, test, thresh_range):
    thresholds = thresh_range
    f1_scores = []

    for threshold in thresholds:
        anomaly_labels = detect_anomalies_with_kmeans(data, y, test, anomaly_threshold=threshold)
        
        # Calculate F1 score
        f1 = f1_score(y_binary_test, anomaly_labels)  # Assuming y_true contains true labels for test data
        f1_scores.append(f1)

    # Plotting
    plt.figure(figsize=(8, 5))
    plt.plot(thresholds, f1_scores, marker='o', linestyle='-')
    plt.title('F1 Score vs. Threshold')
    plt.xlabel('Threshold')
    plt.ylabel('F1 Score')
    plt.grid(True)
    plt.show()
```

## K-Means Cluster:

```python
def kmeans_classifier_train_test(train_data, test_data, num_clusters):
    # Train KMeans on the training data
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    train_labels = kmeans.fit_predict(train_data)
    
    # Calculate the count of different classes within each cluster in the training set
    unique_train_labels = np.unique(train_labels)
    train_label_counts = {}
    for label in unique_train_labels:
        mask = (train_labels == label)
        train_label_counts[label] = np.bincount(y_binary_train[mask])
        
    #print(train_label_counts)
    # Assign label to each cluster based on majority class in the training set
    cluster_labels = {}
    for label, counts in train_label_counts.items():
        majority_class = np.argmax(counts)
        cluster_labels[label] = majority_class

    # Predict labels for the test set based on clusters from the training set
    test_labels = kmeans.predict(test_data)
    predicted_labels = [cluster_labels[label] for label in test_labels]

    return np.array(predicted_labels)

def evaluate_clusters(train_data, test_data, y_train, y_test, max_clusters):
    f1_scores = []

    for i in range(2, max_clusters + 1):
        predicted_labels = kmeans_classifier_train_test(train_data, test_data, i)
        f1 = f1_score(y_test, predicted_labels)
        f1_scores.append(f1)
    
    # Plotting F1 scores for different numbers of clusters
    plt.figure(figsize=(8, 6))
    plt.plot(range(2, max_clusters + 1), f1_scores, marker='o', linestyle='--')
    plt.xlabel('Number of Clusters')
    plt.ylabel('F1 Score')
    plt.title('F1 Score for Different Numbers of Clusters')
    plt.xticks(range(2, max_clusters + 1))
    plt.grid(True)
    plt.show()
    print(np.argmax(f1_scores) + 2)
```

## SVM:

```python
from sklearn.svm import OneClassSVM

def train_and_predict_svm1(train_data, test_data, kernel='rbf', nu=0.1):
    # Train One-Class SVM (SVM-1) on the training data
    svm1 = OneClassSVM(kernel=kernel, nu=nu)
    normal_data = train_data[y_binary_train == 0]
    svm1.fit(normal_data)
    
    # Predict novelties/anomalies in the test data
    predictions = svm1.predict(test_data)
    transformed_predictions = [0 if pred == 1 else 1 for pred in predictions]
    
    return transformed_predictions

def train_and_evaluate_kernels(train_data, test_data, kernels, nu=0.1):
    f1_scores = []

    for kernel in kernels:
        svm1 = OneClassSVM(kernel=kernel, nu=nu)
        normal_data = train_data[y_binary_train == 0]
        svm1.fit(normal_data)
        predictions = svm1.predict(test_data)
        transformed_predictions = [0 if pred == 1 else 1 for pred in predictions]
        f1 = f1_score(y_binary_test, transformed_predictions, average='weighted') 
        f1_scores.append(f1)

    # Plotting F1 scores for different kernels
    plt.figure(figsize=(8, 6))
    plt.bar(kernels, f1_scores, color='skyblue')
    plt.xlabel('Kernel')
    plt.ylabel('F1 Score')
    plt.title('F1 Score for Different Kernels in SVM-1')
    plt.ylim(0, 1)  # Setting y-axis limit for F1 score range (0 to 1)
    plt.show()
```

### Data Norm PCA Probs:

```python
kernels = ['linear', 'rbf', 'poly', 'sigmoid']
```

```python
train_and_evaluate_kernels(d_norm_pca_probs, d_norm_pca_probs_test, kernels)
```

```python
pred_i = train_and_predict_svm1(d_norm_pca_probs, d_norm_pca_probs_test, kernel='rbf')
conf_matrix = metrics.confusion_matrix(y_binary_test, pred_i)

# Define labels for interpretation
labels = ['Normal', 'Anomoly']

# Create a heatmap for the confusion matrix using Matplotlib
plt.figure(figsize=(10, 8))
heatmap = plt.imshow(conf_matrix, cmap='Blues')
plt.colorbar(heatmap)

# Show ticks and labels
plt.xticks(ticks=range(len(labels)), labels=labels)
plt.yticks(ticks=range(len(labels)), labels=labels)

# Set labels and title
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')

# Add annotations to the heatmap
for i in range(len(labels)):
    for j in range(len(labels)):
        plt.text(j, i, conf_matrix[i, j], ha='center', va='center')

plt.show()
# Convert classification report to a DataFrame
print(metrics.classification_report(y_binary_test, pred_i, target_names=labels))
print(f"F1_Score = {f1_score(y_binary_test, pred_i)}")
```
